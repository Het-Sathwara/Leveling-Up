```python
grade = [65,65,47,36,48,39,43,56,78]

#Adding a grade
grade.append(96)
#finding average grade
average =  sum(grade) / len(grade)
print(f"Average Is: {average}")
#Finding highest and lower

high = max(grade)
low = min(grade)

print(f"Highest Score: {high}")
print(f"Lowest Score: {low}")
```

```python
print("Welcome To The Quiz")

play = input("Do You Want To Play? ").lower()

if play != "yes":
    quit()
print("Okay! Let's Play :)")
score = 0

answer = input("What Does CPU Stand For? ").lower()
if answer == "central processing unit":
    print('Correct!')
    score += 1
else:
    print('Incorrect!')

answer = input("What Does RAM Stand For? ").lower()
if answer == "random access memory":
    print('Correct!')
    score += 1
else:
    print('Incorrect!')

answer = input("What Does GPU Stand For? ").lower()
if answer == "graphical processing unit":
    print('Correct!')
    score += 1
else:
    print('Incorrect!')

answer = input("What Does HDD Stand For? ").lower()
if answer == "hard disk drive":
    print('Correct!')
    score += 1
else:
    print('Incorrect!')

print(f"Your final score is {score}.")
```

```python
import random 

range = input("Type A Number: ")

if range.isdigit():
    range = int(range)

    if range <= 0:
        print("Type A Number Bigger Than 0")
        quit()
else:
    
    print('please type a number nest time')
    quit()

r = random.randrange(0, range)
while True:
    guess = input("Make A Guess")

    if guess == r:
        print("you guessed the right number")
        break
    else:
        print("Wrong Guess:(")

    print(f"Actual Number Was: {r}")
```
